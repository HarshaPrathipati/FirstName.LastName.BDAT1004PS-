```python
import pandas as pd
import random
```

```python
s1 = pd.Series([random.randint(1,4) for i in range(100)])
s2 = pd.Series([random.randint(1,3) for i in range(100)])
s3 = pd.Series([random.randint(10000,30000) for i in range(100)])
```

```python
df1 = pd.DataFrame({'s1':s1,'s2':s2,'s3':s3})
df1
```

```python
df1.rename(columns = {'s1':'bedrs','s2':'bathrs','s3':'price_sqr_meter'})
```
